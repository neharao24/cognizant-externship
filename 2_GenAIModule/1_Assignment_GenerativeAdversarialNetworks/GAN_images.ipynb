```python
import torch
from pytorch_pretrained_biggan import BigGAN
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load pretrained BigGAN model
model = BigGAN.from_pretrained('biggan-deep-256')

# Create a class vector for golden retriever
class_vector = torch.zeros(1, 1000)
class_vector[0, 207] = 1

# Generate random latent vector (noise)
latent_vector = torch.randn(1, 128)     # 128 is the size of the input vector for BigGAN

# Generate image
with torch.no_grad():
    output = model(latent_vector, class_vector, 0.4)

# Convert the tensor to a displayable image
generated_image = output.squeeze(0).cpu().numpy()
generated_image = (generated_image.transpose(1, 2, 0) + 1) * 127.5  # Height, Width, Channels for matplotlib
generated_image = generated_image.astype('uint8')

img = Image.fromarray(generated_image)
display(img)




# Experiment with different latent vectors

# Image 2 with a new random latent vector
latent_vector2 = torch.randn(1, 128)

with torch.no_grad():
    output2 = model(latent_vector2, class_vector, 0.4)
    
generated_image2 = output2.squeeze(0).cpu().numpy()
generated_image2 = (generated_image2.transpose(1, 2, 0) + 1) * 127.5
generated_image2 = generated_image2.astype('uint8')

img2 = Image.fromarray(generated_image2)
display(img2)



# Image 3 with a new random latent vector
latent_vector3 = torch.randn(1, 128)

with torch.no_grad():
    output3 = model(latent_vector3, class_vector, 0.4)

generated_image3 = output3.squeeze(0).cpu().numpy()
generated_image3 = (generated_image3.transpose(1, 2, 0) + 1) * 127.5
generated_image3 = generated_image3.astype('uint8')

img3 = Image.fromarray(generated_image3)
display(img3)
```
